```python
sns.countplot(df['OverTime'])

# Set the size of the figure
fig = plt.gcf()
fig.set_size_inches(5, 5)

# Set title
plt.title('OverTime')

# Increase x and y axis values
plt.xticks(fontsize=12)  # Adjust fontsize as needed
plt.yticks(fontsize=12)  # Adjust fontsize as needed

plt.show()
```

```python
sns.countplot(df['MaritalStatus'])
fig=plt.gcf()
fig.set_size_inches(5,5)
plt.title('MaritalStatus')

```

```python
sns.countplot(x='BusinessTravel', data=df)

fig = plt.gcf()
fig.set_size_inches(10, 10)

plt.title('Business Travel')

plt.show()
```

```python
sns.catplot(x="OverTime",y="Age",kind="swarm",data=df)
```

```python
counts = df['TotalWorkingYears'].value_counts()

# Set the size of the figure
plt.subplots(figsize=(30, 10))

# Plot the counts against unique values
sns.barplot(x=counts.index, y=counts.values)

# Set title and labels
plt.title('Count of Total Working Years')
plt.xlabel('Total Working Years')
plt.ylabel('Count')

plt.show()
```
